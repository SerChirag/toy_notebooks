```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate a sine wave
def generate_sine_wave(frequency, amplitude, phase, duration, sample_rate):
    t = np.arange(0, duration, 1/sample_rate)
    y = amplitude * np.sin(2 * np.pi * frequency * t + phase)
    return t, y

# Parameters for three sine waves
frequency1, amplitude1, phase1 = 2, 1.0, 0
frequency2, amplitude2, phase2 = 4, 0.5, np.pi/2
frequency3, amplitude3, phase3 = 8, 0.25, 0


# Parameters for the signal
duration = 2.0  # seconds
sample_rate = 100  # samples per second

# Generate three sine waves
t1, y1 = generate_sine_wave(frequency1, amplitude1, phase1, duration, sample_rate)
t2, y2 = generate_sine_wave(frequency2, amplitude2, phase2, duration, sample_rate)
t3, y3 = generate_sine_wave(frequency3, amplitude3, phase3, duration, sample_rate)

# Create a deterministic mixture of the three sine waves
signal = y1 + y2 + y3
signal2 = y1 + y2 


# Plot the individual sine waves and the mixed signal
plt.figure(figsize=(10, 6))

plt.subplot(5, 1, 1)
plt.plot(t1, y1, label='Sine Wave 1')
plt.legend()

plt.subplot(5, 1, 2)
plt.plot(t1, y2, label='Sine Wave 2')
plt.legend()

plt.subplot(5, 1, 3)
plt.plot(t1, y3, label='Sine Wave 3')
plt.legend()

plt.subplot(5, 1, 4)
plt.plot(t1, signal, label='Mixed Signal')
plt.legend()

plt.subplot(5, 1, 5)
plt.plot(t1, signal2, label='Mixed Signal 2')
plt.legend()


plt.tight_layout()
plt.show()
```

```python
full = np.stack([t1,signal]).T
half = np.stack([t1,signal2]).T
one = np.stack([t1,y1]).T
```

```python
plt.plot(full[:,0], full[:,1], label='Mixed Signal')
```
